:**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,100)
y = x*2
z = x**2

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,y, color="blue")
ax1.plot(x,z, color="red")
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('main plot')


ax2.plot(x,y, color="blue")
ax2.plot(x,z, color="red")
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('insert plot')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)


fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)


ax[0].plot(x, y, color='blue', linestyle='-', linewidth=2, label='y = sin(x)')
ax[0].set_title("(x, y) Plot")
ax[0].set_ylabel("y values")
ax[0].grid(True)
ax[0].legend()


ax[1].plot(x, z, color='red', linestyle='--', linewidth=3, label='z = cos(x)')
ax[1].set_title("(x, z) Plot")
ax[1].set_xlabel("x values")
ax[1].set_ylabel("z values")
ax[1].grid(True)
ax[1].legend()


plt.tight_layout()
plt.show()
```
