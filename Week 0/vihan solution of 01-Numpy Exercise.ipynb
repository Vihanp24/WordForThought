## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
import numpy as np

matrix = np.arange(1, 26).reshape(5, 5)


print(matrix)
```

```python
import numpy as np

full_matrix = np.arange(1, 26).reshape(5, 5)

subarray = full_matrix[2:, 1:]

print(subarray)
```

```python
import numpy as np

full_matrix = np.arange(1, 26).reshape(5, 5)

subarray = full_matrix[3:4, 4:]

print(subarray)
```

```python
import numpy as np

full_matrix = np.arange(1, 26).reshape(5, 5)

subarray = full_matrix[1:2, 0:2]

print(subarray)
```

```python
import numpy as np

full_matrix = np.arange(1, 26).reshape(5, 5)

subarray = full_matrix[4:, 0:]

print(subarray)
```

```python
import numpy as np

full_matrix = np.arange(1, 26).reshape(5, 5)

subarray = full_matrix[3:, 0:]

print(subarray)
```
